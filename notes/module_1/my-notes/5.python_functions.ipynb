## Functions

```python
def list2string(lst):
    string = ' '.join(lst)
    return string

to_string = ['John', 'was', 'a', 'man', 'of', 'many', 'talents']

list2string(to_string)
```

```python
new_list = ['Ana', 'will', 'find', 'an', 'amazing', 'job']
```

```python
list2string(new_list)
```

### Global vs. Local Variables

it is important to keep track of which variables are global and which variables are local. 

- Global variables: maintain their value outside the functions.
- Local variables: maintain their value within the function where they are defined

```python
a = 9

def multiply(number, multiplier = 2):
    b = number * multiplier
    return b

c = multiply(a)

print(a, c)
```

### Applying Functions to Data Frames

```python
import pandas as pd
```

```python
data = pd.read_csv('../week_1/data/vehicles_messy.csv')
```

```python
data.head()
```

```python
def get_means(df):
    numeric = df._get_numeric_data()
    means = pd.DataFrame(numeric.mean()).reset_index()
    means.columns = ['Column', 'Mean']
    return means
```

```python
mean_df = get_means(data)

mean_df
```
