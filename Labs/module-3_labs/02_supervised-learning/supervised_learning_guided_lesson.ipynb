### Preparing the data

1. pick only the columns that we decided to keep
2. separate the data to predictor (X) and response(y) variables.

```python
kickstarter_variables = kickstarter_classify[["usd_goal_real", "backers", "main_category", "duration", "currency"]]
kickstarter_y = pd.get_dummies(data = kickstarter_classify["state"], drop_first = True)
```

```python
kickstarter_x = pd.get_dummies(data = kickstarter_variables, columns = ["main_category", "currency"], drop_first = True)

kickstarter_x.head()
```

3. separate the data into test and train sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(kickstarter_x, kickstarter_y, test_size = 0.2)
```

### Creating the model

we will use logistic regression, but random fores or SVM are also good choices for our problem

```python
# generate model

from sklearn.linear_model import LogisticRegression
```

```python
ks_model = LogisticRegression().fit(X_train, y_train)
```

To evaluate this model we will use a confusion matrix, so we can see how many were correctly classified and how many incorrectly. 

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred_test = ks_model.predict(X_test)
confusion_matrix(y_test, y_pred_test)
```

#### Interpreting the matrix:

1. The first number is the no of observations correctly classified as 0 (or failure).
2. The second are all entries incorrectly clissified as 1 (success).
3. The third is the count of observations incorrectly classified as 0 (failure).
4. The last contains the count of observations correctly classified as 1 (success).


```python
from sklearn import metrics
```

```python
y_pred_proba = ks_model.predict_proba(X_test)[::, 1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)

auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.plot(fpr, tpr);
```

```python
auc
```

The area under the curve is over 0.95. This is a decent number, but maybe we could do better with a different algorithm. 

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ks_rf = RandomForestClassifier().fit(X_train, y_train)
```

```python
# confusion matrix

y_pred_test_rf = ks_rf.predict(X_test)
confusion_matrix(y_test, y_pred_test_rf)
```

```python
# ploting the ROC curve

y_pred_proba_rf = ks_rf.predict_proba(X_test)[::, 1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba_rf)
auc = metrics.roc_auc_score(y_test, y_pred_proba_rf)

plt.plot(fpr, tpr);
```
