```python
import pandas as pd
import numpy as np
```

```python
korea_alc = pd.read_csv("data/korea_alcohol_cons.csv")
usa_alc = pd.read_csv("data/usa_alcohol_cons.csv")
russia_alc = pd.read_csv("data/russia_alcohol_cons.csv")
```

```python
countries_all = pd.concat([korea_alc, usa_alc, russia_alc]).reset_index(drop = True)
```

```python
countries_all
```

```python
countries_all.drop(columns = "Unnamed: 0")
```

```python
countries_all.to_csv("data/alcohol_per_capita.csv")
```
