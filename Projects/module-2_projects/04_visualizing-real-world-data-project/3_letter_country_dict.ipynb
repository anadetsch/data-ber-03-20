```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://laendercode.net/en/3-letter-list.html"
```

```python
laender = BeautifulSoup(requests.get(url).content, "html.parser")
```

```python
land_list = laender.find_all("tr")
```

```python
for i in range(1,len(land_list)):
    land_list[i]= land_list[i].find_all("td")
```

```python
dictionary = {}
```

```python
parse = [land_list[i] for i in range(1,len(land_list))]
```

```python
for items in parse:
    dictionary[items[0].text]=items[1].text
```

```python
dictionary
```

```python
dictionary.to_json("../data/3_letter_country_dict.json")
```
