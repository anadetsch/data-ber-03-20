```python
import pandas as pd
import numpy as np
```

```python
korea = pd.read_csv("data/korea_alcohol_cons.csv")
japan = pd.read_csv("data/japan_alcohol_cons.csv")
russia = pd.read_csv("data/russia_alcohol_cons.csv")
usa = pd.read_csv("data/usa_alcohol_cons.csv")
```

```python
korea = korea.rename(columns = {"Year" : "year", "Alcohol per Capita" : "alcohol per capita"})
korea.head()
```

```python
usa = usa.rename(columns = {"Year" : "year", "Alcohol per Capita" : "alcohol per capita"})

usa.head()
```

```python
countries_all = pd.concat([korea, usa, russia, japan]).reset_index(drop = True)
```
